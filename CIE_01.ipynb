```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars.csv')
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
auto_count=(df['am']==0).sum()
manual_count=(df['am']==1).sum()
labels=['Automatic','Manual']
count=[auto_count,manual_count]
plt.bar(labels, count, color=['orange','blue'])
plt.title('transmission type of cars')
plt.xlabel('types of car')
plt.ylabel('frequency')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=df['wt']
y=df['mpg']
plt.scatter(df['wt'],df['mpg'],color='orange')
plt.title("car weight vs mpg")
plt.xlabel("carsweiht ")
plt.ylabel("miles per gallon")
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
auto=(df['am']==0).sum()
manual=(df['am']==1).sum()
plt.bar(['Automatic','Manual'], [auto, manual], color=['orange','blue'])
plt.title("transmission of car")
plt.ylabel("frequency")
plt.grid(axis='y')
plt.show()
```
